```python
#pairplot-seaborn
#importing packages
import seaborn
import matplotlib.pyplot as plt
#loading dataset using seaborn
df = seaborn.load_dataset('tips')
#df.head()
print(df.to_string())

seaborn.pairplot(df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('tips')
selected_vars=['total_bill','tip']
sns.pairplot(df,vars=selected_vars)
plt.show()
```

```python
#adding ahue color to a seaborn pairplot
import seaborn as sns
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
seaborn.pairplot(df,hue='size')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df=sns.load_dataset('tips')
sns.pairplot(df,hue='size',palette="husl")
plt.show()
```

```python
#different plotkind

import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
sns.pairplot(df,diag_kind='kde')
plt.show
```

```python
#adjusting plot kind

sns.pairplot(df,kind='reg')#reg is used for future prediction
plt.show()
```

```python
#controlling the markers
sns.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
#ADVACENCED CUSTOMIZATION WITH SEABORN
g=sns.pairplot(df,hue='day')
g.fig.suptitle("pairplot of Tips Datasets",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
